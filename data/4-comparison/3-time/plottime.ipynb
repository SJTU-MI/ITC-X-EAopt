```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

def read_log(path):
    
    x_values = []
    y_values = []
    with open(path, 'r') as file:
        for line in file:
            data = line.strip().split()
            x_values.append(float(data[0]))
            y_values.append(float(data[1]))

    return np.array(x_values), np.array(y_values)

GAfile_path = r'./1GA/finalmean.log'
PSOfile_path = r'./2PSO/finalmean.log'
SAfile_path = r'./3SA/finalmean.log'
BOfile_path = r'./4BO/finalmean.log'
MCTSfile_path = r'./5MCTS/finalmean.log'
RSfile_path = r'./6RS/finalmean.log'

GAx, GAy = read_log(GAfile_path)
PSOx, PSOy = read_log(PSOfile_path)
SAx, SAy = read_log(SAfile_path)
BOx, BOy = read_log(BOfile_path)
MCx, MCy = read_log(MCTSfile_path)
RSx, RSy = read_log(RSfile_path)

plt.figure(figsize=(5, 3.8))

sns.lineplot(x=GAx[50:], y=GAy[50:]*2, label='GA', linewidth=3, color='blue')
sns.lineplot(x=PSOx[50:], y=PSOy[50:], label='PSO', linewidth=3, color='green')
sns.lineplot(x=SAx[50:], y=SAy[50:], label='SA', linewidth=3, color='red')
sns.lineplot(x=BOx, y=BOy*4, label='BO', linewidth=3, color='grey')
sns.lineplot(x=MCx, y=MCy, label='MCTS', linewidth=3, color='black')
sns.lineplot(x=RSx[50:], y=RSy[50:], label='RS', linewidth=3, color='pink')

legend = plt.legend(loc='upper left', borderpad=0.5, prop={'weight':'bold'}, bbox_to_anchor=(0.05, 0.98))
plt.xlim(50, 1000)
plt.yscale('log')
plt.gca().spines['top'].set_linewidth(2)
plt.gca().spines['right'].set_linewidth(2)
plt.gca().spines['bottom'].set_linewidth(2)
plt.gca().spines['left'].set_linewidth(2)
plt.tick_params(axis='both', which='major', width=2, length=6)
plt.xticks(fontsize=14,fontweight='bold')
plt.yticks(fontsize=14,fontweight='bold')

plt.xlabel('Search counts', fontsize=14, fontweight='bold')
plt.ylabel('Optimization time (s)', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.savefig('Coretime.jpg', dpi=800)
plt.show()
```
