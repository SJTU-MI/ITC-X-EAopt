```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

dfmax1 = pd.read_csv(r'../1-SG-max/0.01.csv')
dfmax3 = pd.read_csv(r'../1-SG-max/0.03.csv')
dfmax5 = pd.read_csv(r'../1-SG-max/0.05.csv')
dfmin1 = pd.read_csv(r'../2-SG-min/0.01.csv')
dfmin3 = pd.read_csv(r'../2-SG-min/0.03.csv')
dfmin5 = pd.read_csv(r'../2-SG-min/0.05.csv')

colorslight = ['#FFA07A', '#40E0D0', '#6495ED']
colorsdark = ['#FF6347', '#20B2AA', '#4668a6']
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(6, 5), dpi=300)

##################################################max1
sns.boxplot(
    data = dfmax1,
    width = 0.5,
    showfliers = False,
    whis = 0.2,
    boxprops = dict(color = colorslight[0],),    
    capprops = dict(color = colorsdark[0],),   
    medianprops = dict(color = colorsdark[0],),    
    ax = ax1    
)
means = dfmax1.mean()
ax1.plot(
    means.index,
    means, 
    marker = 'o', 
    color = colorsdark[0], 
    label = '1%', 
    markersize = 5, 
    markerfacecolor = colorsdark[0], 
    markeredgecolor = 'black', 
    linestyle = 'dashed'
)

##################################################max3
sns.boxplot(
    data = dfmax3,
    width = 0.4,
    showfliers = False,
    whis = 0.2,
    boxprops = dict(color = colorslight[1],),
    capprops = dict(color = colorsdark[1],),    
    medianprops = dict(color = colorsdark[1],),
    ax = ax1    
)
means = dfmax3.mean()
ax1.plot(
    means.index,
    means, 
    marker = 'o', 
    color = colorsdark[1], 
    label = '3%', 
    markersize = 5, 
    markerfacecolor = colorsdark[1], 
    markeredgecolor = 'black', 
    linestyle = 'dashed'
)

##################################################max5
sns.boxplot(
    data = dfmax5,
    width = 0.3,
    showfliers = False,
    whis = 0.2,
    boxprops = dict(color = colorslight[2],),
    capprops = dict(color = colorsdark[2],),    
    medianprops = dict(color = colorsdark[2],),   
    ax = ax1    
)
means = dfmax5.mean()
ax1.plot(
    means.index,
    means, 
    marker = 'o', 
    color = colorsdark[2], 
    label = '5%', 
    markersize = 5, 
    markerfacecolor = colorsdark[2], 
    markeredgecolor = 'black', 
    linestyle = 'dashed'
)
ax1.set_xticks([])
ax1.set_ylim(320, 400)
ax1.legend(loc = 'lower right')


##################################################min1
sns.boxplot(
    data = dfmin1,
    width = 0.5,
    showfliers = False,
    whis = 0.2,
    boxprops = dict(color = colorslight[0],),
    capprops = dict(color = colorsdark[0],),
    medianprops = dict(color = colorsdark[0],),
    ax = ax2    
)
means = dfmin1.mean()
ax2.plot(
    means.index,
    means, 
    marker = 'o', 
    color = colorsdark[0], 
    label = '1%',
    markersize = 5, 
    markerfacecolor = colorsdark[0], 
    markeredgecolor = 'black', 
    linestyle = 'dashed'
)

##################################################min3
sns.boxplot(
    data = dfmin3,
    width = 0.4,
    showfliers = False,
    whis = 0.2,
    boxprops = dict(color = colorslight[1],),
    capprops = dict(color = colorsdark[1],),
    medianprops = dict(color = colorsdark[1],),
    ax = ax2    
)

means = dfmin3.mean()
ax2.plot(
    means.index,
    means, 
    marker = 'o', 
    color = colorsdark[1], 
    label = '3%', 
    markersize = 5, 
    markerfacecolor = colorsdark[1], 
    markeredgecolor = 'black', 
    linestyle = 'dashed'
)

##################################################min5
sns.boxplot(
    data = dfmin5,
    width = 0.3,
    showfliers = False,
    whis = 0.2,
    boxprops = dict(color = colorslight[2],),
    capprops = dict(color = colorsdark[2],),
    medianprops = dict(color = colorsdark[2],),    
    ax = ax2   
)

means = dfmin5.mean()
ax2.plot(
    means.index,
    means, 
    marker = 'o', 
    color = colorsdark[2], 
    label = '5%', 
    markersize = 5, 
    markerfacecolor = colorsdark[2], 
    markeredgecolor = 'black', 
    linestyle = 'dashed'
)

ax2.set_ylim(175, 225)

for spine in ax1.spines.values():
    spine.set_linewidth(2)
    
for spine in ax2.spines.values():
    spine.set_linewidth(2)

ax1.tick_params(axis='y', labelsize=10, width=2)
ax2.tick_params(axis='y', labelsize=10, width=2)
ax2.tick_params(axis='x', labelsize=10, width=2)
ax2.set_xlabel('Population size ($N$)', fontsize=12, weight='bold')
fig.text(0.02, 0.5, 'σ (MW/m²K)', va='center', rotation='vertical', fontsize=12, weight='bold')

plt.savefig('1-SG-pop.png')
plt.show()
```
